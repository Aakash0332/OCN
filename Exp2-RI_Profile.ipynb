```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n1 = 1.5
delta = 0.01
a = 30
a1 = 125
```

```python
n2 = n1*math.sqrt(1-2*delta)
```

```python
r = np.array(np.linspace(0,a1,a1+1))
n = 0*r
```

```python
for i in range (len(r)):
  if i<a:
    n[i] = n1
  else :
    n[i] = n2

plt.plot(r,n)
plt.title('Step index')
plt.xlabel('radius')
plt.ylabel('n(r)')
```

```python
for alpha in range(1,11):
  for i in range (len(r)):
    if i<a:
      n[i] = float(n1*math.sqrt(1-2*delta*(r[i]/a)**alpha))
    else :
      n[i] = n2
  plt.plot(r,n)
  plt.title('Graded index')
  plt.xlabel('radius')
  plt.ylabel('n(r)')
```
