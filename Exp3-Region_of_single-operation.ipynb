```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a = 3.6
n1 = 1.447
n2 = 1.442
na = math.sqrt(n1**2 - n2**2)
```

```python
lc = ((2*3.14*a*na)/2.405)
lc
```

```python
l = np.arange(0.8,1.8,0.1)
```

```python
v = 0*l
v = ((2*3.14*a*na)/l)
print(v)
```

```python
# Add a label at the midpoint using plt.annotate
plt.annotate('single mode', xy=(lc+0.2, 2.405-0.2) )

plt.plot(l,v)
plt.title('V-number vs wavelength for core diameter')
plt.xlabel('wavelength lambda')
plt.ylabel('V-number')
plt.axhspan(2.405, v[9], color='green', alpha=0.3, lw=0)
plt.axvspan(1.7, lc, color='red', alpha=0.3, lw=0)
```
