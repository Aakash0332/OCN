```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
c = 3*(10**5)
lambda1 = 850
lambda2 = 1300
```

```python
M1 = (1/(c*lambda1))*0.025
M2 = (1/(c*lambda2))*0.004
print(M1)
print(M2)
```

```python
siglam = np.arange(10,100,10)
```

```python
sig_m1 = siglam*M1
sig_m2 = siglam*M2
print(sig_m1)
print(sig_m2)
```

```python
plt.plot(siglam,sig_m1)
plt.plot(siglam,sig_m2)
plt.xlabel('spectral width(Sigma-lambda)')
plt.ylabel('rms pulse broadnening due to material dispertion(s/nm-km)')
plt.title('material dispertion')
plt.legend(["850","1300"],loc='upper left')
```
